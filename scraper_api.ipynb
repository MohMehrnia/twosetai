```python
import requests


scraperapi_url = 'https://api.scraperapi.com/'
payload = {
    'api_key': api_key, 'url': 'https://vcnewsdaily.com/archive.php', 'output_format': 'markdown',
}
response = requests.get(scraperapi_url, params=payload)
print(response.text)
```

```python
import re

markdown_text = response.text

# Regex pattern to match "access/" URLs
urls = re.findall(r'\(access/[^\)]+\)', markdown_text)

# Remove the surrounding parentheses
clean_urls = [url[1:-1] for url in urls]

# Output the list
clean_urls[:3]
```

```python
homepage = 'https://vcnewsdaily.com/'

payload = {
    'api_key': api_key, 'url': homepage + clean_urls[0], 'output_format': 'markdown',
}
response = requests.get(scraperapi_url, params=payload)
print(response.text)
```

```python
markdown_text = response.text

# Regex pattern to match URLs containing 'archivedate.php?...'
matches = re.findall(r'\(archivedate\.php\?[^)]+\)', markdown_text)

# Clean up the parentheses around the URLs
archivedate_urls = [url[1:-1] for url in matches]

# Print results
archivedate_urls
```

```python
homepage = 'https://vcnewsdaily.com/'

archivedate_urls = []

for c_url in clean_urls:
    print(f"Processing {c_url}")
    payload = {
        'api_key': api_key, 'url': homepage + c_url, 'output_format': 'markdown',
    }
    response = requests.get(scraperapi_url, params=payload)
    markdown_text = response.text

    # Regex pattern to match URLs containing 'archivedate.php?...'
    matches = re.findall(r'\(archivedate\.php\?[^)]+\)', markdown_text)
    
    # Clean up the parentheses around the URLs
    archivedate_urls.extend([url[1:-1] for url in matches])
    
```

```python
from IPython.display import display, Markdown
from google.genai import Client
from google.genai import types
from pydantic import BaseModel
import httpx

class Article(BaseModel):
  url: str


gemini_api_key = "API_KEY"
client = Client(api_key=gemini_api_key)

model_id =  "gemini-2.0-flash"
article_urls = []

for i, url in enumerate(archivedate_urls[:100]):
    print(f"Processing {i + 1} url: {url}")
    payload = {
    'api_key': api_key, 'url': homepage + 'access/' + url, 'output_format': 'markdown',
    }
    response = requests.get(scraperapi_url, params=payload)
    llm_response = client.models.generate_content(
        model=model_id,
        contents=[response.text, "extract ONLY all the urls of the news articles in this page. return a JSON list"],
        config={
            'response_mime_type': 'application/json',
            'response_schema': list[Article],
        },
        
    )
    article_urls.extend(llm_response.parsed)
```
